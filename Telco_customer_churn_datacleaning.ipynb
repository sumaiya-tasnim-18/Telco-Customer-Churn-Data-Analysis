# Data Cleaning

### Dataset: Telco_customer_churn.csv

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Telco_customer_churn.csv")
```

```python
df.info()
```

```python
df.dtypes
```

```python
duplicate_rows = df.duplicated().sum()
print("\nNumber of Duplicate Rows:", duplicate_rows)

duplicates = df[df.duplicated()]
print("\nDuplicate Rows Preview:")
print(duplicates.head())
```

```python
missing_values = df.isnull().sum()
print("Missing Values in Each Column:")
print(missing_values)

print("\nTotal Missing Values:", df.isnull().sum().sum())
```

Churn Reason is only filled for customers who actually churned.
Customers who did not churn naturally have no reason, so the value is empty.

From the dataset:

Total customers = 7043

Missing churn reason = 5174

Customers who churned = 7043 − 5174 = 1869
Meaning:

1869 customers churned

5174 customers stayed

This matches the typical churn distribution of the dataset.

For analysis, Replacing missing values with "No Churn":

```python
df['Churn Reason'] = df['Churn Reason'].fillna('No Churn')
```

Datatype Correction

```python
df['Total Charges'] = pd.to_numeric(df['Total Charges'], errors='coerce')
```

```python
df['Total Charges'].dtype
```

Consistency Detection

```python
for col in df.columns:
    print(f"\nColumn: {col}")
    print(df[col].unique())
```
